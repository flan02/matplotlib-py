# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

# Basic Graph

```python


x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your graph
plt.figure(figsize=(5,3), dpi=100) # 300 pixels per inch

opts = {'label': '2x', 'color': 'blue', 'linestyle':'--', 'linewidth': 1, 'marker': '.'} # Dictionarie

#plt.plot(x, y, label='2x', color='#ab30ff') # Shows type of graph generated
plt.plot(x, y, **opts) # Expand the dict and add each key as param

## Select intervals we want to plot points at
x2 = np.arange(0,4.5,0.5)
#print(x2)

# Shorthand
# fmt = '[color][marker][line]'
plt.plot(x2[:6], x2[:6]**2, 'r', label='x²', linewidth= 1) # shorthand
plt.plot(x2[5:], x2[5:]**2, 'r--', label='x² proj', linewidth= 1) # shows a line projection

# Keyword argument notation
plt.title('Our first graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

plt.xlabel('X axis', fontdict={'fontname': 'Arial'})
plt.ylabel('Y axis', fontdict={'fontname': 'Arial'})

plt.xticks([0,1,2,3,4]) # resize x,y graph coordenates
#plt.yticks([0,2,4,6,8,10])

plt.legend()

print(os.getcwd()) # current directory
filename = 'my_first_graph.png'
if not os.path.exists(filename):
    plt.savefig(filename, dpi=300) # Save the graph in the current directory. 300 dpi best quality
    print(f"Image saved as {filename}")
else:
    print(f"The image '{filename}' already exists. It doesn't save again.")

# if os.path.exists("mi_grafico.png"):
#     os.remove("mi_grafico.png")

plt.show() # Shows only the graph
```

# Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
# bars[0].set_hatch('/') # Paints the area of the first bar
# bars[1].set_hatch('O') 
# bars[2].set_hatch('*') 


plt.figure(figsize=(6,4), dpi=100) # 300 pixels per inch

plt.show()
```
