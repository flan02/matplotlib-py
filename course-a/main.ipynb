# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

# Basic Graph

```python


x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your graph
plt.figure(figsize=(5,3), dpi=100) # 300 pixels per inch

opts = {'label': '2x', 'color': 'blue', 'linestyle':'--', 'linewidth': 1, 'marker': '.'} # Dictionarie

#plt.plot(x, y, label='2x', color='#ab30ff') # Shows type of graph generated
plt.plot(x, y, **opts) # Expand the dict and add each key as param

## Select intervals we want to plot points at
x2 = np.arange(0,4.5,0.5)
#print(x2)

# Shorthand
# fmt = '[color][marker][line]'
plt.plot(x2[:6], x2[:6]**2, 'r', label='x²', linewidth= 1) # shorthand
plt.plot(x2[5:], x2[5:]**2, 'r--', label='x² proj', linewidth= 1) # shows a line projection

# Keyword argument notation
plt.title('Our first graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

plt.xlabel('X axis', fontdict={'fontname': 'Arial'})
plt.ylabel('Y axis', fontdict={'fontname': 'Arial'})

plt.xticks([0,1,2,3,4]) # resize x,y graph coordenates
#plt.yticks([0,2,4,6,8,10])

plt.legend()

print(os.getcwd()) # current directory
filename = 'my_first_graph.png'
if not os.path.exists(filename):
    plt.savefig(filename, dpi=300) # Save the graph in the current directory. 300 dpi best quality
    print(f"Image saved as {filename}")
else:
    print(f"The image '{filename}' already exists. It doesn't save again.")

# if os.path.exists("mi_grafico.png"):
#     os.remove("mi_grafico.png")

plt.show() # Shows only the graph
```

## Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

## Histogram

```python
bins = [40,50,60,70,80,90,100] # Shows players overall from 40 to 100

plt.figure(figsize=(8,5), dpi=100)

a = plt.hist(fifa.Overall, bins=bins, color='#663399')
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill level (Overall)')
plt.title('Distribution of Player Skills in FIFA 2018')

filename = 'histogram.png'
if not os.path.exists(filename):
    plt.savefig(filename, dpi=300) # Save the graph in the current directory. 300 dpi best quality
    print(f"Image saved as {filename}")
else:
    print(f"The image '{filename}' already exists. It doesn't save again.")

plt.show()
```

## Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

print('players who use left foot:', left)
print('players who use right foot:', right)

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['red', 'green']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%') # list of numbers [left, right]

plt.title('Foot Preference of FIFA Players')

plt.show()
```

## Pie Chart #2

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot') # Change the color palette of the actual graph

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight] # Removes "lbs" from each value and converts to integer

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

filename = 'fifa_foot_preference.png'
if not os.path.exists(filename):
    plt.savefig(filename, dpi=300) # Save the graph in the current directory. 300 dpi best quality
    print(f"Image saved as {filename}")
else:
    print(f"The image '{filename}' already exists. It doesn't save again.")
    
plt.show()
```

## Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

# We are extracting overall for all players in each team.
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')

filename = 'team_comparison.png'
if not os.path.exists(filename):
    plt.savefig(filename, dpi=300) # Save the graph in the current directory. 300 dpi best quality
    print(f"Image saved as {filename}")
else:
    print(f"The image '{filename}' already exists. It doesn't save again.")
    
plt.show()
```
